## 3. Grouping Data for Summary Analysis

So far, filtering helped us answer:
- Which rows are relevant?

However, filtering still returns **individual rows**.

In many scenarios, we want **summary-level answers**, such as:
- What is the average price per fuel type?
- How does mileage vary across categories?

For this, we use **grouping**.

### 3.1 The Split–Apply–Combine Pattern

Grouping in Pandas follows a three-step process:

1. **Split** the data into groups based on a key  
2. **Apply** a function to each group  
3. **Combine** the results into a new DataFrame

> All groupby operations we see next follow this exact pattern.
